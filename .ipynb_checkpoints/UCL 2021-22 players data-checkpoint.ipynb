## Number of goals scored per club

```python
goals_per_club = gl.groupby('club', as_index=False)['goals'].sum().sort_values(by='goals', ascending=False)
sns.set(rc={'figure.figsize':(20,5)})
plt.xticks(rotation='vertical')
p=sns.barplot(x="club", y="goals", data=goals_per_club)
plt.xlabel("Team", fontsize=20)
plt.ylabel("Goals", fontsize=20)
p.axes.set_title("\nTotal Goals Scored by Teams\n",fontsize=30);
```

## Average playing time per player per club

```python
avg_playing_per_club = ks.groupby('club', as_index=False)['minutes_played'].mean().sort_values(by='minutes_played',ascending=False)
sns.set(rc={'figure.figsize':(20,5)})
plt.xticks(rotation='vertical')
plt.yticks(np.arange(0,600,50))
p=sns.barplot(x='club', y='minutes_played', data=avg_playing_per_club)
plt.xlabel("Team", fontsize=20)
plt.ylabel("Minutes", fontsize=20)
p.axes.set_title("\nAverage playing time per player per Team\n",fontsize=30);
```

- Since Real Madrid is finalist and played 2 extra times, average playing time per player should be higher, but it is still higher than the another finalist Liverpool (4th). So, the reason can be that Real Madrid used less players in UCL, so each player player had to player longer than usual.

## Number of players in each team

```python
avg_playing_per_club = ks.groupby('club').size().reset_index(name='count').sort_values(by='count', ascending=False)
sns.set(rc={'figure.figsize':(20,5)})
plt.xticks(rotation='vertical')
plt.yticks(np.arange(0,34,2))
p=sns.barplot(x='club', y='count', data=avg_playing_per_club)
plt.xlabel("Team", fontsize=20)
plt.ylabel("Players", fontsize=20)
p.axes.set_title("\nNumber of players in each team\n",fontsize=30);
```

- That's right! Real Madrid used 3 players less than Liverpool 

## Top 10 scorers

```python
goals_per_minute = ks['goals'] / ks['minutes_played']
```

```python
ks['goals_per_90'] = goals_per_minute * 90
```

```python
top_scorers = ks.sort_values(by='goals', ascending=False).head(10)
sns.set(rc={'figure.figsize':(20,5)})
plt.xticks(rotation='vertical')
plt.yticks(range(0,16,1))
p=sns.barplot(x='player_name', y='goals', data=top_scorers)
plt.xlabel("Player", fontsize=20)
plt.ylabel("Goals", fontsize=20)
p.axes.set_title("\nTop Scorers\n",fontsize=30);
```

## Goals Per Match

```python
players = ks[ks['match_played'] > 6]
top_scorers_per_match = players.sort_values(by='goals_per_90', ascending=False).head(10)
sns.set(rc={'figure.figsize':(20,5)})
plt.xticks(rotation='vertical')
p=sns.barplot(x='player_name', y='goals_per_90', data=top_scorers_per_match)
plt.xlabel('Player', fontsize=20)
plt.ylabel('Goals per Match', fontsize=20)
p.axes.set_title('\nTop Scorers Per Match\n', fontsize=30)
```

- Benzema is the top scorers of the UCL with 15 goals
- Haller (3rd) has the higest scoring rate per match

## Most Assist Mades

```python
top_assists = ks.sort_values(by='assists', ascending=False).head(10)
sns.set(rc={'figure.figsize':(20,5)})
plt.xticks(rotation='vertical')
plt.yticks(range(0,10,1))
p=sns.barplot(x='player_name', y='assists', data=top_assists)
plt.xlabel("Player", fontsize=20)
plt.ylabel("assists", fontsize=20)
p.axes.set_title("\nMost Assists\n",fontsize=30);
```

Only Sane & Mbappe appears in both scorers & assists lists

Real Madrid has the most goal scoring player and the second most goal assists player

##  Goals Contribution (Goals + Assists)

```python
goals_contribution = ks.sort_values(by='goals_contribution', ascending=False).head(10)
sns.set(rc={'figure.figsize':(20,5)})
plt.xticks(rotation='vertical')
plt.yticks(range(0,17,1))
p=sns.barplot(x='player_name', y='goals_contribution', data=goals_contribution)
plt.xlabel("Player", fontsize=20)
plt.ylabel("Contributed in Goals ", fontsize=20)
p.axes.set_title("\nTop 10 Players Cotribute to Goals\n",fontsize=30);
```

## Temas by Goals

```python
teams = gl.groupby('club')['goals'].agg('sum').reset_index().sort_values('goals', ascending=False)
sns.set(rc={'figure.figsize':(20,5)})
plt.xticks(rotation='vertical')
p=sns.barplot(x='club', y='goals', data=teams)
plt.xlabel("Teams", fontsize=20)
plt.ylabel("Goals", fontsize=20)
p.axes.set_title("\nTemas by Goals\n",fontsize=30);
```

## Goal Types Analysis (How Scored)

```python
goal_sum = pd.DataFrame(gl.sum(numeric_only=True)).T[['right_foot', 'left_foot', 'headers', 'others']]
goal_sum = goal_sum.T
plt.pie(goal_sum[0], autopct='%1.f%%')
plt.title('Goal Types Analysis (How Scored)', fontsize=30)
plt.axis('equal')
plt.legend(goal_sum.index)
plt.show()
```

- Almost half of he goals are scored by right foot.

## Goal Types Analysis (From Where Scored)

```python
goal_sum = pd.DataFrame(gl.sum(numeric_only=True)).T[["inside_area","outside_areas","penalties"]]
goal_sum = goal_sum.T
plt.pie(goal_sum[0], autopct='%1.f%%')
plt.title('Goal Types Analysis (From Where Scored)', fontsize=30)
plt.axis('equal')
plt.legend(goal_sum.index)
plt.show()
```

- Most goals are scored from inside the box.
- Outside of and penalties are about same.

## Goal Types Analysis (Player Position)

```python
goal_sum = gl.groupby('position').sum()['goals']
plt.pie(goal_sum, autopct='%1.f%%')
plt.title('Goals scored from each postion', fontsize=30)
plt.axis('equal')
plt.legend(goal_sum.index)
plt.show()
```

```python
sns.set(rc={'figure.figsize':(15,5)})
plt.xticks(fontsize=15)
p=sns.countplot(x=gl['position'])
plt.xlabel("Postion", fontsize=20)
plt.ylabel("Goals", fontsize=20)
p.axes.set_title("\nGoals per players position\n",fontsize=30)
```

- Although that midfielders players who scored goals are more than forwards but forwards scored more goals

```python
teams_fouls_committed = dsp.groupby('club')['fouls_committed'].agg('sum').reset_index().sort_values(by='fouls_committed', ascending=False)
sns.set(rc={'figure.figsize':(20,5)})
p=sns.barplot(x='club', y='fouls_committed', data=teams_fouls_committed)
plt.xticks(rotation='vertical')
plt.xlabel("Team", fontsize=20)
plt.ylabel("Fouls", fontsize=20)
p.axes.set_title("\nTeams by Fouls Committing\n",fontsize=30);
```

## Players Who Received The Most Yellow Cards

```python
yellow_players = dsp[dsp['yellow'] > 2]
yellow = yellow_players.sort_values(by='yellow', ascending=False)
sns.set(rc={'figure.figsize':(20,5)})
p=sns.barplot(x='player_name', y='yellow', data=yellow)
plt.xticks(rotation='vertical')
plt.yticks(np.arange(0, 5, 1))
plt.xlabel("Player", fontsize=20)
plt.ylabel("Yellow Cards", fontsize=20)
p.axes.set_title("\nPlayers Who Received The Most Yellow Cards\n",fontsize=30);
```

## Players Who Received Red Cards

```python
red_players = dsp[dsp['red'] > 0]
red = red_players.sort_values(by='red', ascending=False)
sns.set(rc={'figure.figsize':(20,5)})
p=sns.barplot(x='player_name', y='red', data=red)
plt.xticks(rotation='vertical')
plt.yticks(np.arange(0, 3, 1))
plt.xlabel("Player", fontsize=20)
plt.ylabel("Red Cards", fontsize=20)
p.axes.set_title("\nPlayers Who Received Red Cards\n",fontsize=30);
```

## Card per Team (Yellow & Red)

```python
teams_cards = dsp.groupby('club')['yellow', 'red'].agg('sum').reset_index().sort_values(by=['yellow', 'red'], ascending=False)
sns.set(rc={'figure.figsize':(20,5)})
p=sns.barplot(x='club', y='yellow', data=teams_cards)
p=sns.barplot(x='club', y='red', data=teams_cards)
plt.xticks(rotation='vertical')
plt.xlabel("Team", fontsize=20)
plt.ylabel("Cards", fontsize=20)
p.axes.set_title("\nTeams by Cards\n",fontsize=30);
```
